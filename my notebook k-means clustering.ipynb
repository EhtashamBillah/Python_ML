```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2,centers=4, cluster_std=1.8,random_state=101)
```

```python
data # THIS IS A TUPLE
```

```python
data[0]
```

```python
plt.scatter(data[0][:,0],data[0][:,1], c= data[1]) # data[1] contains the cluster in the tuples. so we coloring agst the clusters
```

# K Means clustering

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
fig , (ax1,ax2) = plt.subplots(nrows = 1,ncols = 2, sharey=True, figsize = (10,6))
sns.set_style('darkgrid')

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_,cmap ='Dark2',edgecolor='k',linewidths=0.3,alpha=0.8)

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1],cmap='Dark2',edgecolor='k',linewidths=0.3,alpha=0.8)
```
