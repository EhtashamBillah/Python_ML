```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
#print(cancer['DESCR'])
```

```python
df = pd.DataFrame(cancer['data'],columns = cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scale = StandardScaler()
```

```python
scale.fit(df)
```

```python
scaled_data = scale.transform(df)
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
type(x_pca)
```

# Visualizing PCA

```python
sns.set_style('darkgrid')
plt.figure(figsize=(10,6))
plt.scatter(x_pca[:,0],x_pca[:,1],s=25,c=cancer['target'],cmap='Set1', edgecolor='k',alpha = 0.5)
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
# pca.components_
```

```python
# taking Principal components into dataframe
```

```python
df_comp = pd.DataFrame(data = pca.components_, columns = cancer['feature_names'])
```

```python
df_comp.head() # rows are the two principal components
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df_comp,cmap='plasma')
# shows the correaltion between original variables and principal components
```
